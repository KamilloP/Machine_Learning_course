# Task 1
We have two populations Blue (privileged) and Red (unprivileged), with the Blue population being 9 times larger than the Red population.

Individuals from both populations are requesting to attend XAI training to improve competency in this important area. Number of places is limited. The administrators of the training have decided to give priority to enrolling individuals who may need this training in the future, although unfortunately it is difficult to predict who will benefit.

The decision rule adopted:
1. In the Red group, half of the people will find the skills useful in future and half will not. Administrators randomly allocate 50% of people to training.
2. in the Blue group, 80% of people will find the training useful in future and 20% will not, although of course it is not known who will find it useful. The administrators have built a predictive model based on user behaviour in predicting for whom it will be useful and whom will not. The model has the following performance:


| Blue                     	| Will use XAI 	| Will not use XAI 	| Total  |
|--------------------------	|--------------	|------------------	|------- |
| Enrolled in training     	| 60           	| 5               	| 65     |
| not enrolled in training 	| 20            | 15               	| 35     |
| Total                    	| 80           	| 20               	| 100    |


Task: Calculate the Demographic parity, equal opportunity and predictive rate parity coefficients for this decision rule.

Starred task: How can this decision rule be changed to improve its fairness?

## Solution
Let the random variable $A$ represent ethnicity.
We denote that a person belongs to the Red group (unprivileged) by 
$A=r$, and to the Blue group (privileged) by $A=b$.

Let the random variable $\hat{Y}$ represent the decision to grant a student permission to attend the XAI course.
We define $\hat{Y}=1$ as granting permission, which is the preferred outcome.

Let the random variable $Y$ represent whether the student will find the course useful in the future.
We define $Y=1$ as the student finds the training useful.

***Demographic parity | Statistical parity***

Definition:
$$
P(\hat{Y}=1| A=b) \overset{?}{=} P(\hat{Y}=1| A=r)
$$
$$
\hat{Y} \perp A ?
$$
Our model results:
$$
P(\hat{Y}=1| A=b) = \frac{65}{100} = 65\%
$$
$$
P(\hat{Y}=1| A=r) = 50\%
$$
in particular
$$
\frac{P(\hat{Y}=1| A=b)}{P(\hat{Y}=1| A=r)} = \frac{65}{50} = 130\% > 125\% = \frac{5}{4}
$$
Rule of 4/5 is broken.

***Equal opportunity***

Definition:
$$
P(\hat{Y}=1|A=b, Y=1) \overset{?}{=} P(\hat{Y}=1|A=r, Y=1)
$$
Our model results:
$$
P(\hat{Y}=1|A=b, Y=1) = \frac{60}{80} = 75\%
$$
$$
P(\hat{Y}=1|A=r, Y=1) = 50\%
$$
In particular
$$
\frac{P(\hat{Y}=1|A=b, Y=1)}{P(\hat{Y}=1|A=r, Y=1)} = 150\% > 125\%
$$

***Predictive rate parity***

Definition:
$$
P(Y=1|A=b, \hat{Y}=1) \overset{?}{=} P(Y=1|A=r, \hat{Y}=1)
$$
$$
P(Y=1|A=b, \hat{Y}=0) \overset{?}{=} P(Y=1|A=r, \hat{Y}=0),
$$
which is equivalent to
$$
Y \perp A|\hat{Y}?
$$
Our model results:

***Positive predictive parity***
$$
P(Y=1|A=b, \hat{Y}=1) = \frac{60}{65} = \frac{12}{13} = 92.3076923...\%
$$
$$
P(Y=1|A=r, \hat{Y}=1) = 50\%
$$
In particular
$$
\frac{P(Y=1|A=b, \hat{Y}=1)}{P(Y=1|A=r, \hat{Y}=1)} = \frac{24}{13} = 184.6153846...\% > 125\%
$$
***Negative predictive parity***
$$
P(Y=1|A=b, \hat{Y}=0) = \frac{20}{35} = \frac{4}{7} = 57.1428571...\%
$$
$$
P(Y=1|A=r, \hat{Y}=0) = 50\%
$$
In particular
$$
\frac{P(Y=1|A=b, \hat{Y}=0)}{P(Y=1|A=r, \hat{Y}=0)} = \frac{8}{7} = 1,142857143...\% < 125\%
$$
which is good.

*How can this decision rule be changed to improve its fairness?*

The largest disparities appear in positive predictive parity, with equal opportunity showing similar patterns. This indicates that the model itself learns the data distribution reasonably well, but the decision rule applied to the protected group (Red) introduces randomness, which amplifies unfairness.
A potential improvement would be to replace the random decision rule for the Red group with a learned, data-driven classifier. Training a separate model - or adapting the thresholding strategy specifically for this subgroup - could reduce these discrepancies and produce more consistent predictions across groups.

## Solution

### Setup

**Dataset**

In this task we investigate the Adult dataset. It has a medium size and contains a binary target variable.

**Protected attribute**

We select gender (sex) as the protected attribute due to its clear definition and interpretability.
For clarity, we analyze both raw fairness metrics (female, male) and the male/female ratios, which indicate disparity strength.

**Target variable**

The target variable $y$ corresponds to the income attribute.
It is binarized as:

$y = 1 \to$ income >= 50K

$y = 0 \to$ income < 50K

The predicted label is denoted by:

$\hat{y}$  — the model’s prediction for the income class.

This notation is used throughout the evaluation metrics, including accuracy and group fairness measures.

**Models**

**Logistic Regression**

Logistic Regression (LR) is the first baseline model.
It includes L2 regularization, which makes LR more resistant to overfitting.
On the Adult dataset, LR yields comparable accuracy on training and test sets, demonstrating good generalization.

**Random Forest**

Random Forest (RF) as a second baseline model is an ensemble of decision trees and can overfit if trees grow too deep.
To control this, we restrict mainly `max_depth`.

**Logistice Regression on balanced dataset (oversampling)**

We use naive random over-sampling ([RandomOverSampler]). More precisely we generate new samples by randomly sampling with replacement the current available samples.

[RandomOverSampler]: https://imbalanced-learn.org/stable/over_sampling.html#random-over-sampler

### Appendix
The dataset is distributed across three files:
*adult.data*, *adult.names*, and *adult.test*.
The data files (*adult.data* and *adult.test*) do not include column names. The names are provided in *adult.names* and assigned in code:
```
age, workclass, fnlwgt, education, education-num, marital-status,
occupation, relationship, race, sex, capital-gain, capital-loss,
hours-per-week, native-country, income
```

Additionally, the first line of adult.test is:
```
|1x3 Cross validator
```
which must be skipped when loading the file.

Some entries contain missing values encoded as " ?". All rows containing these entries are removed to ensure that both training and test data share exactly the same set of valid categories.

Most features in the Adult dataset are categorical (e.g., workclass, education, race, sex, native-country).
They were transformed using one-hot encoding.

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import RandomOverSampler
````

````python
# Obtain train and test datasets
df1 = pd.read_csv("adult.data", header=None)
df2 = pd.read_csv("adult.test", header=None, skiprows=1)

columns = [
    'age','workclass','fnlwgt','education','education-num',
    'marital-status','occupation','relationship','race','sex',
    'capital-gain','capital-loss','hours-per-week','native-country','income'
]

df1.columns = columns
df2.columns = columns

df1['is_train'] = True
df2['is_train'] = False

df = pd.concat([df1, df2], ignore_index=True)

df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
df.replace("?", "Unknown", inplace=True)
df['income'] = df['income'].str.replace('.', '', regex=False).str.strip()

df['income'] = df['income'].map({'<=50K':0, '>50K':1})
df['sex'] = df['sex'].map({'Female':0, 'Male':1})

# One-hot encoding for other categorical variables
categorical_cols = ['workclass','education','marital-status','occupation','relationship','race','native-country']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

y = df_encoded['income']
X = df_encoded.drop(['income','is_train'], axis=1)

X_train = X[df_encoded['is_train']==True]
X_test  = X[df_encoded['is_train']==False]
y_train = y[df_encoded['is_train']==True]
y_test  = y[df_encoded['is_train']==False]
columns = X_train.columns

X_train = X_train.astype(float).to_numpy()
X_test  = X_test.astype(float).to_numpy()
y_train = y_train.astype(float).to_numpy()
y_test  = y_test.astype(float).to_numpy()

protected_index = columns.get_loc("sex")
````

````python
# 1) Train a model for the selected dataset.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

clf_lr = LogisticRegression(
    random_state=0,
    max_iter=300,
    solver='lbfgs'
)

clf_lr.fit(X_train_scaled, y_train)

print("Accuracy LR (train):", clf_lr.score(X_train_scaled, y_train))
print("Accuracy LR (test):", clf_lr.score(X_test_scaled, y_test))
print("Target (if earns more than 50K) for first 5:", y[:5])
print("Results for first 5:", clf_lr.predict(X_train_scaled[:5,:]))
print("Probabities for first 5:", clf_lr.predict_proba(X_train_scaled[:5,:]))
````

````python
# 2) For the selected protected attribute (age, gender, race) calculate the following 
# fairness coefficients: Statistical parity, Equal opportunity, Predictive parity.

y_p_train1 = clf_lr.predict(X_train_scaled)

def numberOfObservationsForEachCombination(X, y, y_p, protected_id):
    """
    For all combination (s, y, y_) in (Sex X TrueIncome X PredictedIncome) we 
    calculate number of observations that satisfies this combination.
    Parameters:
    1) X - predictors
    2) y - groundtruth income
    3) y_p - predicted income
    Returns:
    3 dimensional tensor T of type int, where T[a,b,c] is the number of observations
    where 
    1) gender is a={0 - Female, 1 - Male}, 
    2) ground truth target is b={0 - <50K, 1 - >=50K},
    3) predicted income is c={0 - <50K, 1 - >=50K}.  
    """
    assert len(X.shape) == 2
    assert len(y.shape) == len(y_p.shape) == 1
    assert X.shape[0] == y.shape[0] == y_p.shape[0]

    T = np.zeros((2, 2, 2), dtype=int)

    for a in range(2):
        mask_a = X[:, protected_id] == float(a)
        y_a = y[mask_a]
        y_p_a = y_p[mask_a]

        for b in range(2):
            mask_b = y_a == float(b)
            y_p_ab = y_p_a[mask_b]

            for c in range(2):
                mask_c = y_p_ab == float(c)
                T[a, b, c] = np.sum(mask_c)
    assert T.sum() == X.shape[0]
    return T


T_train1 = numberOfObservationsForEachCombination(X_train, y_train, y_p_train1, protected_index)
print(T_train1)

def statisticalParity(T):
    return T[:,:,1].sum(axis=1) / T.sum(axis=(1,2))

def equalOpportunity(T):
    return T[:,1,1] / T[:,1,:].sum(axis=1) # axis=1 of T[:,1,:] is actually axis=2 in T

def predictiveParity(T):
    # Actually it is predictive rate parity, so both 
    # positive predictive parity and negative predictive parity.
    return T[:,1,:] / T.sum(axis=1)

print("Statistical parity:", statisticalParity(T_train1))
print("Equal opportunity:", equalOpportunity(T_train1))
print("Predictive parity:", predictiveParity(T_train1))
print("Same, but male/female:")
print("Statistical parity:", statisticalParity(T_train1)[1] / statisticalParity(T_train1)[0])
print("Equal opportunity:", equalOpportunity(T_train1)[1] / equalOpportunity(T_train1)[0])
print("Predictive parity:", predictiveParity(T_train1)[1] / predictiveParity(T_train1)[0])

````

````python
# 3) Train another model (different hyperparameters, feature transformations etc.,
#  different family of models) and see how the coefficients Statistical parity, 
# Equal opportunity, Predictive parity behave for it. Are they different/similar?
clf_rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_leaf=5,
    max_features='sqrt',
    random_state=0,
    n_jobs=-1
) # I did not want overfitting, so I set smaller max_depth and changed min_samples_leaf and max_features.

clf_rf.fit(X_train, y_train)

print("Accuracy RF (train):", clf_rf.score(X_train, y_train))
print("Accuracy RF (test): ", clf_rf.score(X_test, y_test))

T_train2 = numberOfObservationsForEachCombination(X_train, y_train, clf_rf.predict(X_train), protected_id=protected_index)

print("Statistical parity:", statisticalParity(T_train2))
print("Equal opportunity:", equalOpportunity(T_train2))
print("Predictive parity:", predictiveParity(T_train2))
print("Same, but male/female:")
print("Statistical parity:", statisticalParity(T_train2)[1] / statisticalParity(T_train2)[0])
print("Equal opportunity:", equalOpportunity(T_train2)[1] / equalOpportunity(T_train2)[0])
print("Predictive parity:", predictiveParity(T_train2)[1] / predictiveParity(T_train2)[0])
````

````python
# 4) Apply the selected bias mitigation technique (like data balancing) on the first model. 
# Check how Statistical parity, Equal opportunity, Predictive parity coefficients behave after this mittigation.
ros = RandomOverSampler(random_state=0)

# Balanced-X, balanced-y
X_bal, y_bal = ros.fit_resample(X_train, y_train)

scaler2 = StandardScaler()
X_bal_scaled = scaler2.fit_transform(X_bal)
X_test_scaled2 = scaler2.transform(X_test)

clf_lr_bal = LogisticRegression(
    random_state=0,
    max_iter=300,
    solver='lbfgs'
)

clf_lr_bal.fit(X_bal_scaled, y_bal)

print("Accuracy LR-balanced (train):", clf_lr_bal.score(X_bal_scaled, y_bal))
print("Accuracy LR-balanced (test): ", clf_lr_bal.score(X_test_scaled2, y_test))

T_train3 = numberOfObservationsForEachCombination(X_bal, y_bal, clf_lr_bal.predict(X_bal_scaled), protected_index)

print(T_train3)
print("Statistical parity:", statisticalParity(T_train3))
print("Equal opportunity:", equalOpportunity(T_train3))
print("Predictive parity:", predictiveParity(T_train3))
print("Same, but male/female:")
print("Statistical parity:", statisticalParity(T_train3)[1] / statisticalParity(T_train3)[0])
print("Equal opportunity:", equalOpportunity(T_train3)[1] / equalOpportunity(T_train3)[0])
print("Predictive parity:", predictiveParity(T_train3)[1] / predictiveParity(T_train3)[0])
````
